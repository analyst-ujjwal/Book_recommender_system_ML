```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('/content/Books.csv', on_bad_lines = 'skip', encoding='latin-1')
```

```python
df.columns
```

```python
df = df[['ISBN', 'Book-Title', 'Book-Author', 'Year-Of-Publication', 'Publisher']]
```

```python
df.head()
```

```python
df = df.rename(columns={'Book-Title': 'title', 'Book-Author': 'author', 'Year-Of-Publication': 'year', 'Publisher':'publisher'})
```

```python
df.head()
```

```python
users = pd.read_csv('/content/Users.csv')
```

```python
users.rename(columns={'User-ID': 'user_id', 'Location': 'location', 'Age': 'age'}, inplace=True)
```

```python
users.head()
```

```python
ratings = pd.read_csv('/content/Ratings.csv')
```

```python
ratings.head()
```

```python
ratings.rename(columns={'User-ID': 'user_id', 'Book-Rating': 'rating'}, inplace=True)
```

```python
ratings.head()
```

```python
df.shape
```

```python
users.shape
```

```python
ratings.shape
```

```python
x = ratings['user_id'].value_counts()>200
```

```python
y = x[x].index
```

```python
y
```

```python
ratings = ratings[ratings['user_id'].isin(y)]
```

```python
ratings.shape
```

```python
ratings.head()
```

```python
rating_with_books = ratings.merge(df, on='ISBN')
```

```python
rating_with_books
```

```python
rating_with_books.shape
```

```python
number_rating = rating_with_books.groupby('title')['rating'].count().reset_index()
```

```python
number_rating.rename(columns={'rating': 'number_of_ratings'}, inplace=True)
```

```python
final_rating = rating_with_books.merge(number_rating, on= 'title')
```

```python
final_rating.shape
```

```python
final_rating = final_rating[final_rating['number_of_ratings']>=50]
```

```python
final_rating .shape
```

```python
final_rating.drop_duplicates(['user_id', 'title'], inplace = True)
```

```python
final_rating.shape
```

```python
book_pivot = final_rating.pivot_table(columns='user_id', index='title', values='rating')
```

```python
book_pivot.fillna(0, inplace = True)
```

```python
book_pivot
```

```python
from scipy.sparse import csr_matrix
book_sparse = csr_matrix(book_pivot)
```

```python
type(book_sparse)
```

```python
from sklearn.neighbors import NearestNeighbors
model = NearestNeighbors(algorithm='brute')
```

```python
model.fit(book_sparse)
```

```python
distances , suggestion = model.kneighbors(book_pivot.iloc[237,:].values.reshape(1,-1), n_neighbors=6)
```

```python
distances
```

```python
suggestion
```

```python
for i in range(len(suggestion)):
  print(book_pivot.index[suggestion[i]])
```

```python
def recommend_book(book_name):
  id = np.where(book_pivot.index==book_name)[0][0]
  distances , suggestion = model.kneighbors(book_pivot.iloc[id,:].values.reshape(1,-1), n_neighbors=6)
  for i in range(len(suggestion)):
      books = book_pivot.index[suggestion[i]]
      for j in books:
        print(j)
```

```python
recommend_book('Animal Farm')
```
